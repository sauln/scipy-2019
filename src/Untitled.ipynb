```python
import tadasets
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ripser
import persim
from persim import wasserstein
```

```python
sphere = tadasets.dsphere(n=2000)
sphere_noise = tadasets.dsphere(n=2000, noise=0.1)
```

```python
def plot_sphere(sphere):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(sphere[:,0], sphere[:,1], sphere[:,2], s=3, alpha=0.5)
    ax.view_init(elev=18., azim=4)

    plt.tight_layout()

    # ax.axis('square')
    ax.axis('off')
    r = 1.7
    xr = 1.5
    ax.set_xlim3d(-r, r)
    ax.set_ylim3d(-r, r)
    ax.set_zlim3d(-xr, xr)
```

```python
plot_sphere(sphere)
plt.show()
plot_sphere(sphere_noise)
```

```python
dgms_sphere = ripser.ripser(sphere, maxdim=2)
dgms_noise = ripser.ripser(sphere_noise, maxdim=2)
```

```python
d, (mi, D) = wasserstein.wasserstein(dgms_sphere['dgms'][1], dgms_noise['dgms'][1], matching=True)
d2, (mi, D) = wasserstein.wasserstein(dgms_sphere['dgms'][2], dgms_noise['dgms'][2], matching=True)
```

```python
persim.wasserstein_matching(dgms_sphere['dgms'][1], dgms_noise['dgms'][1], mi, labels=["dgm1", "dgm2"], ax=None)
```

```python
persim.wasserstein_matching(dgms_sphere['dgms'][2], dgms_noise['dgms'][2], mi, labels=["dgm1", "dgm2"], ax=None)
```
